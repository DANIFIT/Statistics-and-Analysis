```python
#LOAD LIBRARIES 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#LOAD DATASET
leuanalysisNew1data = pd.read_csv('./leuanalysisNew1.csv')
leuanalysisNew1data.head()
```

```python
#SHOW KEY STATISTICS
leuanalysisNew1data.describe()
```

```python
#Create x and y variables
x = leuanalysisNew1data.drop('REMISS',axis = 1).to_numpy()
y = leuanalysisNew1data['REMISS'].to_numpy()

#Create Training and Test Datasets
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x, y, stratify=y,test_size=0.2,random_state=100)

#Scale the Data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train2 = sc.fit_transform(x_train)
x_test2 = sc.transform(x_test)
```

```python
#Script for SVM and NB
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix


for name,method in [('SVM', SVC(kernel='linear',random_state=100)),
                    ('Naive Bayes',GaussianNB())]: 
    method.fit(x_train2,y_train)
    predict = method.predict(x_test2)
    print('\nEstimator: {}'.format(name))  
    print(confusion_matrix(y_test,predict))
    print(classification_report(y_test,predict,zero_division=0))
```
